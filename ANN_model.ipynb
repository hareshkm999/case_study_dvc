```python
s3 = pd.read_csv('/content/data.csv')
s3.drop("Unnamed: 0", axis=1, inplace=True)
s3.head()
```

```python
# train test split

train, test = train_test_split(s3, test_size=0.3, random_state=42)
test
train_y = train['90_days_repeat']
train_x = train.drop('90_days_repeat', axis=1)
test_y = test['90_days_repeat']
test_x = test.drop('90_days_repeat', axis=1)
```

```python
model = Sequential()
model.add(Dense(12, input_dim=5, activation='relu'))
model.add(Dense(18, activation='relu'))
model.add(Dense(18, activation='relu'))
model.add(Dense(18, activation='relu'))
model.add(Dense(1, activation='sigmoid'))
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
model.fit(train_x, train_y, epochs=150, batch_size=10)
```
